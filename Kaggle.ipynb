```python
# Library imports
import pandas as pd
import plotly.graph_objects as go
from chart_studio.plotly import plot, iplot
```

```python
# Reading timesData from the dataSources folder
timesData = pd.read_csv("/Users/joshua/Documents/gitProjects/Ploty-tutorials/dataSources/timesData.csv")
```

```python
# Preparing dataframe with first 100 rows
df = timesData.iloc[:100,:]

#creating trace1
trace1 = go.Scatter(
                    x = df.world_rank,
                    y = df.citations,
                    mode = "lines",
                    name = "teaching",
                    marker = dict(color = 'rgba(16, 112, 2, 0.8)'),
                    text = df.university_name)

# creating trace 2
trace2 = go.Scatter(
                    x = df.world_rank,
                    y = df.teaching,
                    mode = "lines+markers",
                    name = "teaching",
                    marker = dict(color = 'rgba(80, 26, 80, 0.8)'),
                    text = df.university_name)

data = [trace1, trace2]
layout = dict(title = "Cita and Teachng vs World Rank of top 100 Universities",
              xaxis = dict(title = 'World Rank', ticklen = 5, zeroline=False)
             )
fig = dict(data = data, layout = layout)
fig.show()
```
